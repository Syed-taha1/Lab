```python
# Rotation
from rasterio.plot import show
import rasterio
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
from numpy import expand_dims
from keras.preprocessing.image import img_to_array

image_path =("C:\\Users\\B-Tech\\Downloads\\ALPSRP247640500-H2.2_UA\\HH-ALPSRP247640500-H2.2_UA.tif")
image = rasterio.open(image_path)
data = image.read()
data = data.transpose(1, 2, 0) 
samples = expand_dims(data, 0)
data_generator = ImageDataGenerator(rotation_range=90)
it = data_generator.flow(samples, batch_size=1)
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(330 + 1 + i)
    batch = it.next()
    result = batch[0].astype('uint16')
    plt.imshow(result)

show(image)
```

```python
#Enhancing Color/Color Space
from rasterio.plot import show
import rasterio
from numpy import expand_dims
from keras.preprocessing.image import ImageDataGenerator, img_to_array
import matplotlib.pyplot as plt

image_path = ("C:\\Users\\B-Tech\\Downloads\\ALPSRP247640500-H2.2_UA\\HH-ALPSRP247640500-H2.2_UA.tif")
image = rasterio.open(image_path)

data = image.read()
data = data.transpose(1, 2, 0) 

samples = expand_dims(data, 0)
datagen = ImageDataGenerator(brightness_range=[0.5, 1])
it = datagen.flow(samples, batch_size=1)

plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(330 + 1 + i)
    batch = it.next()
    result = batch[0].astype('uint16')
    plt.imshow(result)

plt.show()
```
